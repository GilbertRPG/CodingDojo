```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.heatmap([[87,16],[17,59]],
            cmap='Blues', 
            annot=True,
            );
accuracyScore = (59+87)/(87+16+17+59) # https://classeval.wordpress.com/introduction/basic-evaluation-measures/#:~:text=Accuracy%20is%20calculated%20as%20the,dataset%20(P%20%2B%20N).
plt.title('Accuracy Score: {:.6f}'.format(accuracyScore))
plt.xlabel('Predicted Label')
plt.ylabel('Actual Label')
plt.show()
```
